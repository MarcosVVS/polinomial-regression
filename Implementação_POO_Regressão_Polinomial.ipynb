```python
import numpy as np
```

```python
# Extraindo amostras
X_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([i[0] ** 2 for i in X_train])
X_test = np.array([[1], [2], [3], [4], [5]])
y_test = np.array([i[0] ** 2 for i in X_test])
```

```python
class PolynomialRegression:

    '''
    Docstring
    '''

    def __init__(self, degree, alpha = 0.01, n_iters = 1000):
        self.degree = degree
        self.alpha = alpha
        self.n_iters = n_iters

    def fit(self, X, y):
        m = X.shape[0]

        # Eleva cada feature aos graus contidos no range
        # O '+1' em self.degree é porque o range vai de 1 a self.degree - 1
        # Colunas de 1
        X_poly = np.hstack([X ** i for i in range(0, self.degree + 1)])

        print(X_poly)
        # Vetor com a quantidade de zeros igual ao número de colunas de X_poly
        print(X_poly.shape)
        self.w_ = np.zeros(self.degree + 1)
        self.b_ = 0

        # 'np.dot(A, B)' é o mesmo que 'A @ B'
        #  O _ evita que uma variável seja criada na memória
        for _ in range(self.n_iters):

            dJdw = np.dot(X_poly.T, ((np.dot(X_poly, self.w_) + self.b_) - y))
            dJdb = (np.dot(X_poly, self.w_) + self.b_) - y

            # Atualização simultânea dos parâmetros
            self.w_ = self.w_ - (1/m) * self.alpha * dJdw
            print(self.w_)
            self.b_ = self.b_ - (1/m) * self.alpha * dJdb
            print(self.b_)

    def predict(self, X):
        X_poly = np.hstack([X ** i for i in range(0, self.degree + 1)])
        y_pred = np.dot(self.w_, X_poly) + self.b_
        return y_pred
```

```python
model = PolynomialRegression(degree = 3, alpha = 0.03, n_iters = 100)
model.fit(X_train, y_train)
```
